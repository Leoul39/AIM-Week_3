```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore',category=pd.errors.DtypeWarning)
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
data=pd.read_csv('C:/Users/abenet/Desktop/Trial/AIM-Week_3/data/dataset.csv')
data.head()
```

```python
df=data.copy()
df.columns 
```

### Cleaning null values

```python
df['VehicleType']=df['VehicleType'].fillna('Passenger Vehicle')
df['Model']=df['Model'].fillna('QUANTUM 2.7 SESFIKILE 16s')
df['cubiccapacity'].fillna(df['cubiccapacity'].mean(),inplace=True)
df['kilowatts'].fillna(df['kilowatts'].mean(),inplace=True)
df['Cylinders'].fillna(df['kilowatts'].mean(),inplace=True)
```

* For the statistical modeling, I used certain columns that I thought were relevant to the claims people ask. The columns used are:
    1. Features about the cars to be insured: `ExcessSelected`, `VehicleType`,`Model`, `cubiccapacity`, `kilowatts`, `RegistrationYear`,`Cylinders`
    2. Features about the client and its location: `PostalCode`,`Gender_new`, `Province`,`SubCrestaZone`
    3. Features about the insurance coverage: `CoverType`,`SumInsured`
    

```python
df_cleaned=df[['ExcessSelected','VehicleType','cubiccapacity','kilowatts','Cylinders','IsVATRegistered',
               'Gender_new','Province','PostalCode','CoverType','LegalType','StatutoryClass','SumInsured','TotalPremium']]
df_cleaned.head()
```

* Since the original data was cleaned, we'll have no null values.

```python
df_cleaned.isnull().sum()
```

```python
sample_size=300000
sample=df_cleaned.sample(n=sample_size,random_state=0)
sample.reset_index(drop=True,inplace=True)
sample.head()
```

```python
df_= pd.get_dummies(data=sample, prefix='is', prefix_sep='_',
                       columns=['ExcessSelected','VehicleType','IsVATRegistered','LegalType',
             'Province','PostalCode','Gender_new','StatutoryClass','CoverType'], drop_first=True, dtype='int8')
df_.head()
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
X=df_.drop('TotalPremium',axis=1)
y=df_['TotalPremium']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
r2_score(y_pred,y_test)
```

```python
feature_importance = pd.Series(model.coef_, index=X_train.columns).sort_values(ascending=False)
feature_importance.head(10)
```

```python
df_['TotalClaims']=df['TotalClaims']
df_d=df_.drop('TotalPremium',axis=1) 
X=df_d.drop('TotalClaims',axis=1)
y=df_d['TotalClaims']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
r2_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_jobs=-1)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
r2_score(y_pred,y_test)
```
